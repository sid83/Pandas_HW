```python
# Function to analyze scores and calculate passing rates
def result_analysis(df):
        mean_math=df['math_score'].agg(np.mean)
        mean_reading=df['reading_score'].agg(np.mean)
        df_pass_math=df[df['math_score']>=70]
        df_pass_read=df[df['reading_score']>=70]
        math_percent=len(df_pass_math)/len(df)*100
        read_percent=len(df_pass_read)/len(df)*100
        overall_percent=(math_percent+read_percent)/2
        return([mean_math,mean_reading,math_percent,read_percent,overall_percent])
        
```
